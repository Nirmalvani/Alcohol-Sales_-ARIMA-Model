#### Let's see the stationarity of the Sales

```python
plt.plot(df)
```

```python
# We can see a trend here. Let's see it mathematically.
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
# Function to check stationarity mathematically
def ad_test(dataset):
    a = adfuller(dataset,autolag="AIC")
    print(f"1. ADF : {a[0]}")
    print(f"2. P-value : {a[1]}")    
    print(f"3. Num of lags : {a[2]}")
    print(f"4. Number of observation used for ADF regression and Critical values calculation : {a[3]}")
    print(f"5. Critical values : ")
    for key,val in a[4].items():
        print("\t",key,":",val)
```

```python
ad_test(df['Sales'])
```

**P- value >0.05** so the mathematically as well the data is non stationary.

Lets make it stationary

```python
df['Sales_difference'] = df['Sales'] - df['Sales'].shift(1)
```

```python
ad_test(df['Sales_difference'].dropna()) # Now it is stationry data.
```

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Plot the ACF
plt.figure(figsize=(10, 5))
plot_acf(df['Sales_difference'].dropna(), lags=100, ax=plt.gca())
plt.title('ACF Plot')
plt.show()
```

```python
# Plot the PACF
plt.figure(figsize=(10, 5))
plot_pacf(df['Sales_difference'].dropna(), lags=60, ax=plt.gca())
plt.title('PACF Plot')
plt.show()
```

```python
# Assuming `sales_stationary` is your stationary time series data
train_size = int(len(df['Sales_difference']) * 0.8)  # Use 80% of data for training
train = df['Sales_difference'][:train_size]
test = df['Sales_difference'][train_size:]
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Assuming 'data' is your stationary time series
model = ARIMA(train, order=(96, 1, 24))  # Replace p, d, q with appropriate values
model_fit = model.fit()
```

```python
# Print model summary
print(model_fit.summary())
```

```python
# Forecast the next points corresponding to the length of the test data
predictions = model_fit.forecast(steps=len(test))

plt.figure(figsize=(10, 6))
plt.plot(test, label="Actual (Test Data)")
plt.plot(predictions, label="Predicted (Forecast)", color='red')
plt.title("Actual vs Predicted")
plt.legend()
plt.show()
```

```python
# Assuming you have the last value before differencing
last_original_value = df['Sales'].iloc[train_size - 1]  # The last value before the test set

# Invert the differencing for predictions
predictions_original_scale = predictions.cumsum() + last_original_value

# You can then plot `predictions_original_scale` against the original `sales` data
```

```python
plt.plot(df['Sales'], label="Actual (Test Data)")
plt.plot(predictions_original_scale, label="Predicted (Forecast)", color='red')
plt.legend()
plt.show()
```

```python
mape = np.mean(np.abs((test - predictions) / test)) * 100
print(f'Mean Absolute Percentage Error (MAPE): {mape}%')
```

## We got 90% accuracy. And visually it is is following the following the trend.
